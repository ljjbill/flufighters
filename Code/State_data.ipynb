```python
# Dependencies
import pandas as pd

# Store filepath in a variable
file_one = "Resources/state_population_data.txt"

# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")


# Show just the header
file_one_df.head()
```

```python
three_field = file_one_df[["State" , "Population estimate" , "Percent of total U.S. population"]].head(100)
three_field.head()
```

```python
three_field.to_csv("Resources/StateTerritory_population_v1.csv", index=False, header=True)
```

```python
# gettting state abbrevation
state_abbrev = "Resources/state_abbrev.csv"

# stroring state abbrev in data frame
# Not every CSV requires an encoding, but be aware this can come up
state_abbrev_df = pd.read_csv(state_abbrev, encoding="ISO-8859-1")


# Show just the header
state_abbrev_df.head()
```

```python
## merging state to get postal code

merge_table_df = pd.merge(three_field, state_abbrev_df, on="State", how="right")

## dropping null value
state_pop_abbrev = merge_table_df.dropna()
state_pop_abbrev
final_state_pop=state_pop_abbrev.rename(columns={'State':'state_name','Postal Code':'State'})
final_state_pop
```

```python
three_field.to_csv("Resources/State_population.csv", index=False, header=True)
```
